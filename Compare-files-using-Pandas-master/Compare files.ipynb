```python
# Import pandas library
import pandas as pd
```

```python
# Load excel file
df1 = pd.read_excel('file1.xlsx')
df1.head(2)
```

```python
# Load excel files
df2 = pd.read_excel('file2.xlsx')
df2.head(2)
```

```python
# merge both files using merge (similar concept like joins)

df_join = df1.merge(right = df2,
                    left_on = df1.columns.to_list(),
                    right_on = df2.columns.to_list(),
                    how = 'outer')
```

```python
# By simply merging the fles we find that after files are merged, we are unable to differentiate the records
df_join
```

```python
# So we have to rename columns in both the files

df1.rename(columns = lambda x : x + '_file1', inplace=True)
df2.rename(columns = lambda x : x + '_file2', inplace=True)
```

```python
# then perform merge functin again

df_join = df1.merge(right = df2,
                    left_on = df1.columns.to_list(),
                    right_on = df2.columns.to_list(),
                    how = 'outer')
```

```python
# Now we can differentiate between records from both the files
df_join
```

```python
# to get records_present_in_df1_not_in_df2:-

records_present_in_df1_not_in_df2 = df_join.loc[df_join[df2.columns.to_list()].isnull().all(axis = 1), df1.columns.to_list()]
```

```python
records_present_in_df1_not_in_df2
```

```python
# to get records_present_in_df2_not_in_df1:-

records_present_in_df2_not_in_df1 = df_join.loc[df_join[df1.columns.to_list()].isnull().all(axis = 1), df2.columns.to_list()]
```

```python
records_present_in_df2_not_in_df1
```

```python
# This is how we can compare 2 files using python pandas library.
```

```python
record_1 = records_present_in_df1_not_in_df2
record_2 = records_present_in_df2_not_in_df1
```

```python
df_new_join = record_1.merge(right=record_2,
                            left_on=record_1.columns.to_list(),
                            right_on=record_2.columns.to_list(),
                            how = "outer")
```

```python
df_new_join
```

```python
df_new_join.to_excel('output.xlsx')
```
